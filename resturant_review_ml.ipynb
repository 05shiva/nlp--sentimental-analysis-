```python
x=df.Review
y=df.Liked_or_not
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
obj_cv=CountVectorizer()
x_train_cv=obj_cv.fit_transform(x_train.values)
```

```python
x_train_np=x_train_cv.toarray()
```

```python
obj_cv.vocabulary_
```

```python
obj_cv.get_feature_names_out()[]
```

```python
obj_cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_cv,y_train)
```

```python
x_test_cv=obj_cv.transform(x_test)
from sklearn.metrics import classification_report
y_pridict=model.predict(x_test_cv)
print(classification_report(y_test,y_pridict))
```
